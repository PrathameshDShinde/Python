```python
Body_Temperature = int(input("Enter Temperature"))

if Body_Temperature > 30 :
    
    print("It is a Hot Day")
    
elif Body_Temperature < 30 :
    
    print("It is Cold Day")
    
else :
    print("Its neither Hot Or Cold Day")
```
